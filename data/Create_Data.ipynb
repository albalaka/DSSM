```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import re
import tensorflow as tf
```

```python
def load_data(name):
    # Data has dimensions (#days, lane ID, time)
    # In total, there are 440 days of data for 963 lanes
    # Measurements were taken every 10 minutes, giving 6 measurements * 24 hours = 144 measurements per lane, per day
  
    # Labels represent day of the week starting from 1 = monday to 7 = sunday
    data = pd.read_csv('PEMS-SF/PEMS_{}'.format(name), sep = ';', header = None).values
    for i in range(data.shape[0]):
        data[i][0] = data[i][0][1:]
        data[i][962] = data[i][962][:-1]
    for i in range(data.shape[0]):
        for j in range(data.shape[1]):
            data[i][j] = data[i][j].split(' ')

    f = open('PEMS-SF/PEMS_{}labels'.format(name), 'r')
    labels = [int(x) for x in re.sub('[\[\]]', '', f.readline()).split(' ')]
    return(np.asarray(data.tolist(), dtype = np.float), np.asarray(labels))

def invert(original):
    # invert a permutation
    return(np.argsort(original))
```

```python
f = open('PEMS-SF/randperm', 'r')
perm = list(re.sub('[\[\]]', '', f.readline()).split(' '))
f.close()
perm = np.asarray(perm, dtype = np.int)-1
inverse_perm = invert(perm)
```

```python
train_X,train_y = load_data('train')
test_X,test_y = load_data('test')
print(test_X.shape, test_y.shape)
tot_X = np.concatenate((train_X,test_X))
tot_y = np.concatenate((train_y, test_y))
print(tot_X.shape,tot_y.shape)
original_ordered_data = tot_X[inverse_perm] # original ordered data (#days,laneID, Time of Day) 
original_ordered_DoW = tot_y[inverse_perm] # Day of Week labels
print(original_ordered_data.shape,original_ordered_DoW.shape)
```
